```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('StudentsPerformance.csv')
```

```python
students_perfomance = students_perfomance.rename(columns = 
                    {
                        'parental level of education':'parental_level_of_education',
                        'test preparation course':'test_preparation_course',
                        'math score':'math_score',
                        'writing score':'writing_score',
                        'reading score':'reading_score'
                    })
```

```python
students_perfomance.head()
```

```python
students_perfomance.groupby('gender', as_index=False).aggregate({'math_score':['mean', 'count', 'median'], 'reading_score':['mean', 'count', 'median']})
```

```python
students_perfomance.groupby('gender')[['math_score', 'reading_score', 'writing_score']].agg([np.mean, np.median])
```

```python
students_perfomance.groupby(['gender', 'race/ethnicity']).math_score.nunique()
```

```python
students_perfomance.sort_values(['gender', 'math_score'], ascending=False).groupby('gender').head()
```

```python
students_perfomance.groupby(['gender', 'race/ethnicity']).aggregate({'math_score':['mean','median']})
```

```python
students_perfomance.groupby('gender').aggregate({'writing_score':'mean'})
```

```python
students_perfomance['total_score'] = students_perfomance.math_score + students_perfomance.writing_score + students_perfomance.reading_score
```

```python
students_perfomance['total_score'] = students_perfomance.filter(like='score').sum(axis=1)
```

```python
students_perfomance = students_perfomance.assign(total_score_log = np.log(students_perfomance.total_score))
```
